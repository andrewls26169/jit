```python
import csv
import json
```

```python
INFILE = "Teleological.csv"
OUTFILE = "hypertree1.js"
```

```python
infile = csv.DictReader(open(INFILE))
lines = list(infile)
print(lines)
```

```python
nodes=[]
for line in lines:
    # update edges depending on what's in each node
    node={"id": line["ID"], "name": line["Name"],
         "data": {"text":line["Text"]},
         "adjacencies":[]}
    nodes.append(node)
for line in lines:
    if line["To"]:
        id_from=line["ID"]
        id_to=line["To"]
        for node in nodes:
            if node["id"] == id_to:
                if line["Type"] == 'R':
                    node["adjacencies"].append({"nodeTo":id_from, 
                                                "data":{"weight" : 1, 
                                                        "$color": "#f00"}})
                else:
                    node["adjacencies"].append({"nodeTo":id_from, 
                                                "data":{"weight" : 1}})
            if node["id"] == id_from:
                if line["Type"] == 'R':
                    node["adjacencies"].append({"nodeTo":id_to, 
                                                "data":{"weight" : 1,
                                                        "$color": "#f00"}})
                else:
                    node["adjacencies"].append({"nodeTo":id_to, 
                                                "data":{"weight" : 1}})
                
nodes
```

```python
print(json.dumps(nodes, indent=4))
```

```python
with open(OUTFILE, 'w') as f:
    with open('hypertree-preamble.txt') as p:
        f.write(p.read())
    json.dump(nodes, f, indent=4) 
    with open('hypertree-postamble.txt') as p:
        f.write(p.read())

        
```
